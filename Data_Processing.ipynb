```python
# I want this
# Merge DataFrames on 'column1' with a left join
left_merge = pd.merge(top_50, most_streamed[['track_name','artist_count','in_spotify_playlists','streams']], on='track_name', how='left')
print("\nLeft Merge:")
left_merge
```

## Removing extra data points

```python
left_merge['track_name'].value_counts()
```

```python
# Sorting by 'other_column' here because we might prefer to keep the row with the max or min in 'other_column'
left_merge = left_merge.sort_values(by=['streams'], ascending=[True])

# Drop duplicates keeping the first occurrence after sorting
# 'key_column' is the column where duplicates are identified
left_merge = left_merge.drop_duplicates(subset=['track_name'], keep='first')
```

```python
left_merge
```

## Replacing NaN for Artist Count

```python
def count_artists_from_track(track_name):
    # Regex to find patterns of additional artists
    pattern = re.compile(r'feat\.|with')
    if pattern.search(track_name):
        # Split by 'feat.' or 'with', and further by '&' to count additional artists
        parts = re.split(r'feat\.|with', track_name)[-1]
        # Further split by '&' and ',', and strip spaces to handle multiple artists correctly
        additional_artists = re.split(r'&|,', parts)
        # Count and strip extra spaces
        artist_count = sum(1 for name in additional_artists if name.strip())
        return artist_count + 1  # Add 1 for the main artist
    return 1  # Default count if no additional artists are mentioned

# Apply the function and fill NaN in 'artist_count'
left_merge['artist_count'] = left_merge.apply(
    lambda row: count_artists_from_track(row['track_name']) if pd.isna(row['artist_count']) else row['artist_count'],
    axis=1
)

left_merge
```
